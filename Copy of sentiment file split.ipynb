```python

df = pd.read_csv("covid-19_vaccine_tweets_with_sentiment.csv", encoding = 'unicode_escape')

df['label2'] = df['label'].apply(lambda x: x-1)

train = df.sample(frac=0.7,random_state =42  )
test = df.loc[~df.index.isin(train.index)]


```
